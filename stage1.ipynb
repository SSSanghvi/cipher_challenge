```python
def get_letter_frequencies_in_string(s):
    letters = list(string.ascii_uppercase)
    counts = {letter: 0 for letter in letters}
    num_letters = 0
    for char in s:
        if char in letters:
            counts[char] += 1
            num_letters += 1
    return {letter: count/num_letters for (letter, count) in counts.items()}
```

```python
letters = list(string.ascii_uppercase)

plaintext_sample_frequencies = get_letter_frequencies_in_string(plaintext_sample)
freqs = [plaintext_sample_frequencies[l] for l in letters]

sns.barplot(x=letters, y=freqs)
plt.title('Letter Frequencies in Plaintext Sample');
```

```python
def count_single_letter_appearances(text):
    counts = dict(zip(string.ascii_uppercase, [0] * 26))
    for word in text.split(' '):
        if len(word) == 1 and word in string.ascii_uppercase:
            counts[word[0]] += 1
    return counts
            
```

```python
counts_single_letter_appearances = count_single_letter_appearances(plaintext_sample)
counts = [counts_single_letter_appearances[k] for k in counts_single_letter_appearances.keys()]

sns.barplot(x=list(string.ascii_uppercase), y=counts)
plt.title('Counts of Single-Letter Words in Plaintext Sample');
```

```python
def count_n_letter_words(n, text):
    counts = {}
    for word in text.split(' '):
        if len(word) == n and word.isalpha():
            if word in counts.keys():
                counts[word] += 1
            else:
                counts[word] = 1
    return counts
```

```python
def substitute_characters(substitutions, ciphertext):
    ciphertext = ciphertext.upper()
    substitution_text = ''
    for letter in ciphertext:
        if letter in substitutions.keys():
            substitution_text += substitutions[letter]
        else:
            substitution_text += letter
    return substitution_text
```
